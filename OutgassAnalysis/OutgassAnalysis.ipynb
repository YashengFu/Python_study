## This is a program to analyze the experimental test data of outgassing rate
### How to obtain raw data: 
* Obtain raw data from RGA software, import the data into .xlsx for analysis.

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import matplotlib.ticker as mtick
```

```python
plt.style.use('./my.mplstyle')

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
materials = ['background','SiPM-support','FR4-forward','Cable-26','Samtec-connector','glue-seal']
#materials = ['background','SiPM-support','Rogers-forward','Cable-26','Samtec-connector','glue-seal']
#materials = ['kapton','teflon','background','FR4-forward','SiPM-support','Rogers-forward','Cable-26','Samtec-connector','glue-seal']
for material in materials: 
    df = pd.read_excel('./data/outgassing.xlsx',engine='openpyxl',sheet_name=material,header=1,usecols=['Oxygen-mol','Time'])
    data=df.head()
    #for i in np.arange(len(df['Time']))
    if material =='background':
        S=df['Oxygen-mol']-df['Oxygen-mol']
        B=df['Oxygen-mol']
        continue
    S = -5*B+S+df['Oxygen-mol']
    plt.plot(df['Time'],df['Oxygen-mol'],label=material)#,color='red')
plt.plot(df['Time'],S,label='ToTal')#,color='red')
plt.legend(loc='upper right',ncol=2,prop={'size': 12})
plt.xlabel('Time(s)',size =20)
plt.ylabel('mol/(L·S)',size =20)
plt.title('Variation of outgassing molor mass over time(Oxygen)',size = 20,pad=20)
plt.tick_params(labelsize=18,grid_linestyle='--')
plt.grid(True)
plt.ylim(1e-14,3e-10)

plt.yscale('log')
plt.savefig('Outgassing-result.pdf')
```
